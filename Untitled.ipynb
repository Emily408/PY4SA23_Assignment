```python
import pandas as pd

df = pd.read_csv("~/Documents/PY4SA23_Assignment/portland_park_trees.csv")

filtered_trees = df[(df["Genus"].isin(["Quercus", "Acer"])) & (df["DBH"] > 50)]
num_filtered_trees = len(filtered_trees)
print(num_filtered_trees) # could do f to add text
```

```python
filtered_trees = df[df["Genus"].isin(["Quercus", "Acer", "Fraxinus"])]

genus_mean_dbh = filtered_trees.groupby("Genus").mean()

highest_mean_genus = genus_mean_dbh.idxmax()["DBH"]

print(highest_mean_genus)
```

```python
acer_genus = df[df["Genus"] == "Acer"]
num_species = acer_genus["Species"].value_counts().shape[0]
print(num_species)
```
